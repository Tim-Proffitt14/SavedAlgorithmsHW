First we import the math folder and set up a Heap class. I used the max heap function from our heapsort problem earlier in the year, and just changed the signs to make it a Min Heap

```python
import math

class TreeNode:
  def __init__(self, value=None):
    self.value = value
    self.children =[]

  def get_value(self):
    return self.value

  def set_value(self, value):
    self.value = value
    
  def get_children(self):
    return self.children
    
  def add_child(self, child_node):
    self.children.append(child_node)

  def remove_child(self, child_node):
    self.children.remove(child_node)



class Heap:
  def __init__(self, A):
    self.max_heap = A
    self.heap_size = len(A)

  def parent(self, i):
    return self.max_heap[math.floor(i / 2)]

  def left(self, i):
    return 2 * i +1

  def right(self, i):
    return 2 * i + 2

  def max_heapify(self, i):
    l = self.left(i)
    r = self.right(i)
    largest = i
    if l < self.heap_size and self.max_heap[l] > self.max_heap[i]:
      largest = l
    if r < self.heap_size and self.max_heap[r] > self.max_heap[largest]:
      largest = r
    if largest != i:
      self.max_heap[i], self.max_heap[largest] = self.max_heap[largest], self.max_heap[i]
      self.max_heapify(largest)
    

  def build_max_heap(self):
    self.heap_size = len(self.max_heap)
    for i in range(self.heap_size // 2 -1, -1, -1):
      self.max_heapify(i)
      print(self.max_heap)
    

  def heapsort(self):
    self.build_max_heap()
    for i in range(len(self.max_heap) -1, 0,-1):
      self.max_heap[i], self.max_heap[0] = self.max_heap[0], self.max_heap[i]
      self.heap_size = self.heap_size - 1
      self.max_heapify(0)
      print(self.max_heap)
  
  
  
  def print_heap(self):
    print(self.max_heap)
```

Now we set up our Node and our graph class, including the Prim's algorithm method. This graph class uses an adjacency matrix and stores the wait of the edges in the correct locations. I also set it up so that it only works for undirected graphs, but that can probably be changed fairly easily.

```python
class Node:
  def __init__(self, cost, parent = None, key = 100000):
    self.cost = cost
    self.parent = parent
    self.key = key

class Graph:
  def __init__(self, graph_dict):
    self.graph_dict = graph_dict

  def add_edge(self, x, y, value):
    self.graph_dict[x][y].cost = value
    self.graph_dict[y][x].cost = value

  def Prims(self,r):
    v = 10
    selector = [False] * v
    edges = 0
    selector[0] = True
    while(edges < v -1):
      minimum = 10000
      x = 0
      y = 0
      for i in range(v):
        if selector[i]:
          for j in range(v):
            if((not selector[j])) and self.graph_dict[i][j].cost:
              if minimum > self.graph_dict[i][j].cost:
                minimum = self.graph_dict[i][j].cost
                x =i
                y = j
      print(f'x: {x + 1}  y: {y +1 }, cost : {minimum}')
      selector[y] = True
      edges += 1
        
```
